```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Read the CSV file Sales YTY Percent Change
csv_path = Path("Resources/Sales YTY Percent Change.csv")
sales_yty_df = pd.read_csv(csv_path)

sales_yty_df
```

```python
state_per_and_month_df = sales_yty_df[["Mon-Yr","CA*"]]
```

```python
# Splitting the Mon-Yr column to have a column for Year and Month
state_per_and_month_df[["year","month"]] = state_per_and_month_df["Mon-Yr"].str.split('-', n=1, expand=True)
state_per_and_month_df
```

```python
state_per_and_month_df["CA*"] = state_per_and_month_df["CA*"].str.replace("%","")
```

```python
state_per_and_month_df.dtypes
```

```python
# Make the values of the California column astype float
state_per_and_month_df["CA*"] = state_per_and_month_df["CA*"].astype("float")
```

```python
# Display dtypes
state_per_and_month_df.dtypes
```

```python
# Find the average percent for each year for Califonia Percent Change
yearly_mean= state_per_and_month_df.groupby("year")["CA*"].mean()
yearly_mean
```

```python
# Line Plot chart of California average percent change through 2018 - 2023
yearly_mean.plot()
plt.title('Average Percent Change of Housing Prices in California (2018 - 2023)')
plt.xticks(rotation=45)
plt.ylabel('Percent Change')
plt.xlabel('Year')
plt.show()
```

```python
state_per_and_month_df = sales_yty_df[["Mon-Yr","CA*"]]

state_per_and_month_df[["year","month"]] = state_per_and_month_df["Mon-Yr"].str.split('-', n=1, expand=True)

state_per_and_month_df.isnull().sum()
```

```python
state_per_and_month_df.dtypes
```

```python
state_per_and_month_df["year"] = state_per_and_month_df["year"].astype("int")
state_per_and_month_df["CA*"] = state_per_and_month_df["CA*"].str.replace("%","").astype("float")

print(state_per_and_month_df.dtypes)

state_per_and_month_df.head()
```

```python
# Find the California Housing Price Percent Change for each year and each month
years_df = pd.DataFrame(columns=["month"])

for year in [23,22,21,20,19,18]:
    temp_df = state_per_and_month_df.loc[state_per_and_month_df["year"]==year, ["CA*", "month"]]
    
    temp_df = temp_df.rename({"CA*": str(year)}, axis='columns')
    
    years_df = temp_df.merge(years_df, how="left")
    
years_df = years_df.loc[:, ["month", "18", "19", "20", "21", "22", "23"]]

years_df
```

```python
# Create a line plot chart with each line representing a year
ax = plt.gca()

years_df.plot(kind='line', x='month', ax=ax)

plt.title("CA Housing Report by Monthly Percentage from 2018 to 2023")
plt.xticks(rotation=90)
plt.show()
```

```python
# Finding the average California Housing Price Percent Change for each year
years_series = state_per_and_month_df.groupby("year")["CA*"].mean()

years_series
```

```python
# Creating a bar graph to show the average percent change for each year
plt.bar(years_series.index.values, years_series.values)

plt.title("CA Housing Report by Yearly Average from 2018 to 2023")
plt.xlabel('year')
plt.ylabel('Average')

plt.show()
```
